#### ***Analysing sales according to the branch***

```python
location_list = sales.groupby('Location')
```

```python
location = [x for x, y in location_list]
location
```

```python
sales.groupby('Location').sum()["Total"]
```

```python
plt.bar(location, sales.groupby('Location').sum()["Total"], color = 'crimson')
plt.grid()
plt.show()
```

#### ***Sales according to location***

```python
plt.pie(sales.groupby('Location').sum()["Total"], labels = location, autopct = '%1.1f%%')
plt.xticks(rotation = 0)
plt.show()
```

##### ***Finding out which branch has more female crowd and which brach has more male***

```python
location_sales =  sales.groupby(['Gender', 'Location']).count()['Invoice ID']
unstacked_sales =  location_sales.unstack(level = 0)
```

```python
unstacked_sales.plot(kind='bar', figsize = (10,6))
plt.xticks(rotation = 0)
plt.grid()
plt.show()
```

##### ***1. Which branch has more members vs which has less members?***
##### ***2. Which branch has highest rating and which has lowest?***

```python
members =  sales.groupby(['Member', 'Location']).count()['Invoice ID']
members
```

```python
members.unstack(level = 0).plot(kind = 'bar', figsize = (10, 6))
plt.xticks(rotation = 0)
plt.grid()
plt.show()
```

```python
rating = sales.groupby('Location').mean()["Rating"]
rating
```

```python
rating.plot(kind = 'bar', color = 'yellow')
plt.xticks(rotation = 0)
plt.grid()
plt.show()
```

#### ***Which city has more female shoppers?***

```python
female_shoppers = sales.groupby(['Gender', 'City']).count()["Invoice ID"]
```

```python
female_shoppers.unstack(level=0)
```

```python
female_shoppers.unstack(level=0).plot(kind = 'bar', figsize = (10,6))
plt.xticks(rotation = 0)
plt.grid()
plt.show()
```

#### ***Who spends more males or females?***

```python
spend = sales.groupby(['Gender']).sum()['Total']
```

```python
spend.plot(kind = 'bar', color = 'cyan')
plt.xticks(rotation = 0)
plt.grid()
plt.show()
```
